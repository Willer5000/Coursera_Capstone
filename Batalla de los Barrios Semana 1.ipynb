4. Foursqare
El siguiente código utiliza la API de Foursquare para buscar restaurantes en Buenos Aires que estén abiertos en el momento de hacer la búsqueda. Se especifican las coordenadas geográficas del centro de la ciudad y se utiliza el parámetro "sort" para ordenar los resultados por distancia. Luego se hace una solicitud HTTP GET a la URL de la API de Foursquare con los parámetros y cabeceras necesarios. La respuesta se convierte en un objeto JSON y se almacena en la variable "data".

```python
import pandas as pd
import requests

url = "https://api.foursquare.com/v3/places/search"

params = {
  	"query": "restaurant",
  	"ll": "-34.603722,-58.381592",
  	"open_now": "true",
  	"sort":"DISTANCE"
}

headers = {
    "Accept": "application/json",
    "Authorization": "fsq3pz4x+pH00OKi1LqAAw2+HLEmHnzJgVb53Qkpow+sZKc="
}

response = requests.request("GET", url, params=params, headers=headers)

data = response.json()
data
```

El siguiente código convierte los resultados en un DataFrame de Pandas que contiene información sobre los restaurantes que cumplen con los criterios de búsqueda especificados.

```python
# Verificar que la clave 'results' existe en los datos devueltos
if 'results' in data:
    # Si existe, convertir los resultados en un dataframe
    results = data['results']
    restaurants = []
    for r in results:
        name = r.get('name', None)
        latitude = r['geocodes']['main'].get('latitude', None)
        longitude = r['geocodes']['main'].get('longitude', None)
        restaurants.append([name, latitude, longitude])
    df_CFBA2 = pd.DataFrame(restaurants, columns=['name', 'latitude', 'longitude'])
else:
    # Si no existe, imprimir mensaje de error
    print("No se encontraron resultados para la búsqueda especificada.")

```
